```python
#What are the total sales attributed to customers over the age of 45?
df_invoice = pd.read_csv("fct_invoice.csv")
df_invoice.rename(columns={'customer_id':'id'}, inplace=True)
df_invoice['sale_value'] = df_invoice['quantity']*df_invoice['price']
df_invoice = df_invoice.merge(df_customer, how='left')
df_invoice_age = df_invoice[df_invoice['age'].notnull()]
above_45 = df_invoice_age['age'] > 45
sales_above45 = df_invoice_age.groupby(above_45)['sale_value'].sum().get(True,0)
print(sales_above45)
```

```python
#How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)
import matplotlib.pyplot as plt
```

```python
#ensuring date data are recognized as date time objects and not strings
df_invoice['invoice_date'] = pd.to_datetime(df_invoice['invoice_date'], dayfirst=True)
df_invoice
```

```python
#sorting
df_invoice = df_invoice.sort_values(by='invoice_date')
df_invoice
```

```python
#creating line chart with sale value across all dates in the dataframe
ax = df_invoice.plot(x='invoice_date', y='sale_value', figsize=(100,6))
print(ax)
```

```python
#had to use something from online here but it decomposes the sales data into trends and seasonality charts
from statsmodels.tsa.seasonal import seasonal_decompose

df_invoice.set_index('invoice_date', inplace=True)

analysis = df_invoice[['sale_value']].copy()


decompose_result_mult = seasonal_decompose(analysis, model="multiplicative", period=12)

trend = decompose_result_mult.trend
seasonal = decompose_result_mult.seasonal
residual = decompose_result_mult.resid

decompose_result_mult.plot();
```

```python
'''
Hard
1. Create a pivot table showing the breakdown of sales across these dimensions, in this order: category, decade age range (e.g., 10-19, 20-29, and so forth).
'''
```

```python
#create function to determine decade age range.
def decade_range(x):
    float(x)
    lower_bound = 0
    upper_bound = 9
    loop = 1
    while loop == 1:
        if lower_bound <= x <= upper_bound:
            
            age_range = str(lower_bound)+'-'+str(upper_bound)
            return age_range
            loop = 0
        else:
            lower_bound = lower_bound +10
            upper_bound = upper_bound +10
```

```python
#apply function to create new column in invoice dataframe
df_invoice = pd.read_csv("fct_invoice.csv")
df_invoice.rename(columns={'customer_id':'id'}, inplace=True)
df_invoice['sale_value'] = df_invoice['quantity']*df_invoice['price']
df_invoice = df_invoice.merge(df_customer, how='left')
df_invoice_age_range = df_invoice[df_invoice['age'].notnull()]
df_invoice_age_range['age'] = df_invoice_age_range['age'].apply(decade_range)
df_invoice_age_range
```

```python
#create pivot table
df_invoice_age_range.pivot_table('sale_value',['category', 'age'], aggfunc='sum')
```
