```python
import requests
import time
import datetime
import matplotlib.pyplot as plt

#function to get current ISS position
def iss():
    iss_url = "https://api.wheretheiss.at/v1/satellites/25544"
    data = requests.get(iss_url).json()
    lat = float(data["latitude"])
    lon = float(data["longitude"])
    time = float(data["timestamp"])
    return {"lat":lat, "lon":lon, "timestamp":time}
        
#use function to get list of data points
points = []
for x in range(110): #110 to account for possible corrupted data returns from API
    points.append(iss())
    time.sleep(10) #wait 10 seconds before getting the next data point from the API
lats = [p["lat"] for p in points]
lons = [p["lon"] for p in points]
times = [p["timestamp"] for p in points]
time1 = times[0]
time1_dateformat = datetime.datetime.utcfromtimestamp(time1)
string_time1_dateformat = time1_dateformat.strftime('%Y-%m-%d %H:%M:%S UTC')

#create graph using points
plt.figure(figsize=(10, 5))
plt.plot(lons, lats, marker='o', markersize=3, color='blue', linewidth=1)
plt.title("ISS Position Every 10 Seconds Since "+string_time1_dateformat)
plt.xlabel("Longitude (°)")
plt.ylabel("Latitude (°)")
plt.grid(True)

# Marking the start and end points
plt.scatter(lons[0], lats[0], color='green', s=50, label='Start')
plt.scatter(lons[-1], lats[-1], color='red', s=50, label='End')
plt.legend()

plt.tight_layout()
plt.show()
```
